```python
from torchvision import models 
import torch 
```

```python
arch = 'resnext101_32x8d'
arch_lst =  ['resnet50', 'resnet101','resnet152', 'resnext50_32x4d', 'resnext101_32x8d']

input = torch.rand((1,3,418,418))
for arch in arch_lst:
    print(arch)
    backbone = models.__dict__[arch](pretrained=True)
    layer0 = backbone.conv1(input)
    print(layer0.shape)
    layer0 = backbone.bn1(layer0)
    print(layer0.shape)
    layer0 = backbone.relu(layer0)
    print(layer0.shape)
    layer0 = backbone.maxpool(layer0)
    print(layer0.shape)
    layer1 = backbone.layer1(layer0)
    print(layer1.shape)
    layer2 = backbone.layer2(layer1)
    print(layer2.shape)
    layer3 = backbone.layer3(layer2)
    print(layer3.shape)
    layer4 = backbone.layer4(layer3)
    print(layer4.shape)

    print('-------------------------')
```
